```python
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

# df_01 = pd.read_excel('C:/Users/YSJ/Desktop/운동일지/운동일지.xlsx', sheet_name='운동 일지', index_col='날짜', usecols='A:E')
df_02 = pd.read_excel('C:/Users/YSJ/Desktop/운동일지/운동일지.xlsx', sheet_name='2분할 계획표 실시', index_col='날짜', usecols='A:I')

# df_01
df_02.fillna('', inplace=True) # index로 지정된 값들은 fillna로 변경 불가능(아마도 dtype이 다르지 않을까). 먼저 날짜를 제외한 부분의 NaN데이터를 지우고 난 다음, 날짜 부분에서는 ffill로 NaT 데이터를 채운다.
df_02.reset_index(inplace=True)
df_02.fillna(method='ffill', inplace=True)
df_02.set_index('날짜', inplace=True)
df_02
```
